# MovieClient Example

The `MovieClient` class allows users to pull a movie from TheMovieDB, and process that movie into a single row dataframe which matches the training data format. Movies are referenced by title, and the API client will attempt to find a movie matching that title.

## Setup

```python
from pprint import pprint

from bechdel_test_predictor import MovieClient
```

```python
client = MovieClient()
```

## Load a movie

```python
the_big_short = client.get_processed_movie(title="The Big Short")
```

You can also get the raw data movie data, stored in the `Movie` class, prior to pre-processing features.
